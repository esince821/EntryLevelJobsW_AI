# Hypothesis 3: Education Requirements

This hypothesis examines whether education requirements are related to experience level.

- Null Hypothesis (H0): Education requirement is independent of experience level.
- Alternative Hypothesis (H1): Education requirement depends on experience level.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

df2 = pd.read_csv("ai_jobs_market_cleaned.csv")
df2.head()

edu_table = pd.crosstab(df2["experience_level"], df2["education_required"])
print(edu_table)
```

The table shows how education requirements are distributed across experience levels. Each experience group includes a mix of all education categories, such as Associate, Bachelor, Master, PhD, and Bootcamp/Self-taught.

There is no clear pattern indicating that higher experience levels consistently require higher degrees. The distribution appears relatively similar across groups, with only small differences.

This suggests that education requirements may not strongly depend on experience level. A Chi-square test is needed to confirm whether this relationship is statistically significant.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.countplot(
    data=df2,
    x="education_required",
    hue="experience_level"
)

plt.title("Education Requirement by Experience Level")
plt.xlabel("Education Requirement")
plt.ylabel("Count")

plt.xticks(rotation=45)
plt.show()
```

### Visualization of Education Requirements

The countplot shows the distribution of education requirements across experience levels.

Consistent with the contingency table, all experience levels include a mix of different education categories. There is no clear pattern where higher experience levels require significantly higher degrees.

Overall, the visualization suggests that education requirements are fairly similar across experience levels.
